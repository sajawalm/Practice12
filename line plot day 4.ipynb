### Import libraries
- seaborn automatically installs these following libraries
*numpy*
*scipy*
*pandas*
*matplotlib*

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set theme
sns.set_theme(style = "ticks", color_codes="True")# load dataset
phool = sns.load_dataset("iris")# Iris means phool
phool
sns.lineplot(x ="sepal_length", y="sepal_width", data =phool) # lineplot hamesha 2 numeric variable k dirmyan banta hai
plt.show()
```

### adding title and set the limits of x and y

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set theme
sns.set_theme(style = "ticks", color_codes="True")# load dataset
phool = sns.load_dataset("iris")# Iris means phool
phool
sns.lineplot(x ="sepal_length", y="sepal_width", data =phool) # lineplot hamesha 2 numeric variable k dirmyan banta hai
plt.title("phoolon ka title")
plt.xlim(3)
plt.ylim(1)
plt.show()
```

### size of figure

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set theme
sns.set_theme(style = "ticks", color_codes="True")# load dataset

# change figure
plt.figure(figsize=(8,6))

phool = sns.load_dataset("iris")# Iris means phool
sns.lineplot(x ="sepal_length", y="sepal_width", data =phool) # lineplot hamesha 2 numeric variable k dirmyan banta hai
plt.title("phoolon ka title")
plt.show()
```
